## Editar un DataFrame

### Agregar una columna

```python
df4['concatenar']= df4['nombre']+df4['nombre']
df4.head(10)
```

```python
indices = df4.index.values
indices
```

```python
lista_tuplas = []
for i in indices:
    lista_tuplas.append((df4["lat"][i],df4["lon"][i]))
```

```python
lista_tuplas[:10]
```

```python
df4['coordenadas']= lista_tuplas
df4.head(10)
```

### Reemplaza valores de una columna

```python
df4['nombre'].replace({'Sevilla':'Sevilla_linea1','Chapultepec':'chapultepec_linea1'})
df4
```

```python
df4['nombre'].replace({'Sevilla':'Sevilla_linea1','Chapultepec':'chapultepec_linea1'}, inplace=True)
df4.head(10)
```

### Cambiar nombre de una columna
