### Lectura de datos csv, json, xls

Uno de los formatos más utilizados en la actualidad para intercambio de datos es CSV (“Comma Separated Values”). 

Este es archivo de texto en los que cada línea contiene una fila de datos con múltiples registros delimitados por un separador. 

En general el separador suele ser la coma, de ahí el nombre del formato. 

Aunque también se pueden utilizan otros caracteres que no suelen estar contenidos en los datos. Por ejemplo, espacios, tabuladores y puntos y coma.

## Editar un DataFrame

### Agregar una columna

```python
df4['concatenar']= df4['nombre']+df4['nombre']
df4.head(10)
```

```python
indices = df4.index.values
indices
```

```python
lista_tuplas = []
for i in indices:
    lista_tuplas.append((df4["lat"][i],df4["lon"][i]))
```

```python
lista_tuplas[:10]
```

```python
df4['coordenadas']= lista_tuplas
df4.head(10)
```

### Reemplaza valores de una columna

```python
df4['nombre'].replace({'Sevilla':'Sevilla_linea1','Chapultepec':'chapultepec_linea1'})
df4
```

```python
df4['nombre'].replace({'Sevilla':'Sevilla_linea1','Chapultepec':'chapultepec_linea1'}, inplace=True)
df4.head(10)
```

### Cambiar nombre de una columna

```python
df4.columns
```

```python
df4.rename(columns={'nombre':'Estación','linea':'Línea','num':'Número de estación', 'coordenadas':'Datos geoespaciales'}, inplace = True)
df4.head(10)
```

### Borrar columnas o filas

```python
df4.drop('lat', axis = 1, inplace = True)
df4.head(10)
```

```python
df4.drop([2,3], axis = 0, inplace = True)
df4.head(10)
```

```python
del df4['lon']
df4.head(10)
```
